```python
pl_link=engl_competitions[engl_competitions['competition']=='Premier League']['comp_href'].tolist()[0]
```

```python
pl_link
```

```python
pl_res=requests.get(pl_link)
pl_page=pl_res.text
pl_doc = BeautifulSoup(pl_page, 'html.parser')
result_href='https://livescores.com' + pl_doc.find('a',{'class':'Bd'})['href']
base_url='https://livescores.com'
Hg=[]
Ag=[]
hT=[]
aT=[]
home_shotsOnTarget=[]
away_shotsOnTarget=[]
home_shotsOffTarget=[]
away_shotsOffTarget=[]
home_fouls=[]
away_fouls=[]
home_corners=[]
away_corners=[]
home_throws=[]
away_throws=[]
h_line=[]
h_form=[]
a_form=[]
a_line=[]
h_subs=[]
a_subs=[]
h_coach=[]
a_coach=[]
htHG=[]
htAG=[]
ref=[]
m_date=[]
m_venue=[]
                
while True:
    result_res=requests.get(result_href)
    result_page=result_res.text
    result_doc=BeautifulSoup(result_page, 'html.parser')
    matches=result_doc.find_all('div',{'class':'tf xf'})
    for match in matches:
        goals=match.find('span',{'class':'Vh'})
        if goals:
            teams=match.find_all('span',{'class':'Sh'})
            hT.append(teams[0].text)
            aT.append(teams[1].text)
            Hg.append(goals.text)
            Ag.append(match.find('span',{'class':'Wh'}).text)
            match_href=base_url+ match.find('a',{'class':'vc'})['href']
            match_res=requests.get(match_href)
            match_page= match_res.text
            match_doc=BeautifulSoup(match_page,'html.parser')
            match_cat=match_doc.find('div',{'class':'He'})
            match_cat_links=match_cat.find_all('a',{'class':'Ie'})
            info_href=base_url+match_cat_links[0]['href']
            stats_href=base_url+match_cat_links[2]['href']
            line_up=base_url+match_cat_links[3]['href']
            info_res=requests.get(info_href)
            info_doc=BeautifulSoup(info_res.text,'html.parser')
            stats_res=requests.get(stats_href)
            stats_doc=BeautifulSoup(stats_res.text,'html.parser')
            line_up_res=requests.get(line_up)
            line_up_doc=BeautifulSoup(line_up_res.text,'html.parser')
            stat_div=stats_doc.find_all('div',{'class':'De'})
            for stat in stat_div:
                stat_name=stat.find('div',{'class':'Ge'}).text
                home_stat=stat.find('span',{'class':'Ae'}).text
                away_stat=stat.find('span',{'class':'Be'}).text
                if stat_name== 'Shots on target':
                    home_shotsOnTarget.append(home_stat)
                    away_shotsOnTarget.append(away_stat)
                elif stat_name== 'Shots off target':
                    home_shotsOffTarget.append(home_stat)
                    away_shotsOffTarget.append(away_stat)
                elif stat_name== 'Corner Kicks':
                    home_corners.append(home_stat)
                    away_corners.append(away_stat)
                elif stat_name== 'Fouls':
                    home_fouls.append(home_stat)
                    away_fouls.append(away_stat)
                elif stat_name== 'Throw ins':
                    home_throws.append(home_stat)
                    away_throws.append(away_stat)
                else:
                    pass
            home_line=line_up_doc.find_all('span',{'class':'Zh'})[0].text
            home_line=re.sub('\D+','', home_line)
            home_line='-'.join(list(home_line))
            h_form.append(home_line)
            away_line=line_up_doc.find_all('span',{'class':'Zh'})[1].text
            away_line=re.sub('\D+','', away_line)
            away_line='-'.join(list(away_line))
            a_form.append(away_line)
            home=line_up_doc.find('div',{'class':'ni'}).text
            home=re.sub('\d+',',',home)
            home=home.split(',')
            h_line.append(home)
            away=line_up_doc.find('div',{'class':'oi'}).text
            away=re.sub('\d+',',',away)
            away=away.split(',')
            a_line.append(away)
            homesubs=[]
            awaysubs=[]
            subs=line_up_doc.find_all('div',{'class':'Ih'})
            for i in range(0,len(subs)-1):
                homesubs.append(subs[i].find('span',{'class':'Ch'}).text)
                awaysubs.append(subs[i].find('span',{'class':'Dh'}).text)
            h_subs.append(homesubs)
            a_subs.append(awaysubs)
            h_coach.append(subs[-1].find('span',{'class':'Ch'}).text)
            a_coach.append(subs[-1].find('span',{'class':'Dh'}).text)
            ht_scores=info_doc.find('div', {'class':'Aj'})
            htHG.append(ht_scores.find('span',{'id':'match-detail__home__score'}).text)
            htAG.append(ht_scores.find('span',{'id':'match-detail__away__score'}).text)
            misc=info_doc.find_all('div', {'class':'kc'})
            ref.append(misc[0].text)
            m_date.append(misc[1].find_all('span')[1].text)
            m_venue.append(misc[1].find_all('span')[0].text)
        else:
            pass
    next_page=result_doc.find('a',{'type':'next'})
    if next_page == None:
        break
    else:
        result_href=base_url+  next_page['href']
        
```

```python
data=pd.DataFrame({'home team': hT,'away team': aT,'FTHG':Hg,'FTAG': Ag,'htHG': htHG, 'htAG':htAG,'home formation': h_form,
                   'away formation': a_form,'home line up': h_line,'away line up':a_line,'home subs': h_subs, 
                   'away subs':a_subs,'home coach': h_coach,'away coach':a_coach,'Referee':ref,'date':m_date,'venue':m_venue,
                   'home shots on target':home_shotsOnTarget,'away shots on target':away_shotsOnTarget,
                   'home shots off target':home_shotsOffTarget, 'away shots off target': away_shotsOffTarget,
                   'home fouls': home_fouls,'away fouls': away_fouls,'home corners': home_corners, 
                   'away corners': away_corners,'home throws': home_throws, 'away throws': away_throws})
```

```python
data
```

```python
data['home line up']=data['home line up'].apply(lambda x: ','.join(x))
data['away line up']=data['away line up'].apply(lambda x: ','.join(x))
data['home subs']=data['home subs'].apply(lambda x: ','.join(x))
data['away subs']=data['away subs'].apply(lambda x: ','.join(x))
```

```python
data.to_csv('premier_league.csv', index=False)
```

```python
df=pd.read_csv('premier_league.csv')
```

```python
df
```

```python
def result(r):
    if (r['FTHG']-r['FTAG']>0):
        return 'H'
    elif (r['FTHG']-r['FTAG']==0):
        return 'D'
    elif (r['FTHG']-r['FTAG']<0):
        return 'A'
    

df['result']=df.apply(lambda r: result(r), axis=1)
    
    
```

```python
def home_pos(r):
    ht_pos=r['htHG']-r['htAG']
    ft_pos=r['FTHG']-r['FTAG']
    if ht_pos==0 and ft_pos==0:
        return 'drew at ht and ft'
    if ht_pos==0 and ft_pos>0:
        return 'drew at ht and won at ft'
    if ht_pos==0 and ft_pos<0:
        return 'drew at ht and lost at ft'
    if ht_pos>0 and ft_pos==0:
        return 'won at ht and drew at ft'
    if ht_pos>0 and ft_pos>0:
        return 'won at ht and won at ft'
    if ht_pos>0 and ft_pos<0:
        return 'won at ht and lost at ft'
    if ht_pos<0 and ft_pos==0:
        return 'lost at ht and drew ft'
    if ht_pos<0 and ft_pos>0:
        return 'lost at ht and won at ft'
    if ht_pos<0 and ft_pos<0:
        return 'lost at ht and ft'

df['home_ht_ft']=df.apply(lambda r: home_pos(r), axis=1)
    
```

```python
def away_pos(r):
    ht_pos=r['htAG']-r['htHG']
    ft_pos=r['FTAG']-r['FTHG']
    if ht_pos==0 and ft_pos==0:
        return 'drew at ht and ft'
    if ht_pos==0 and ft_pos>0:
        return 'drew at ht and won at ft'
    if ht_pos==0 and ft_pos<0:
        return 'drew at ht and lost at ft'
    if ht_pos>0 and ft_pos==0:
        return 'won at ht and drew at ft'
    if ht_pos>0 and ft_pos>0:
        return 'won at ht and won at ft'
    if ht_pos>0 and ft_pos<0:
        return 'won at ht and lost at ft'
    if ht_pos<0 and ft_pos==0:
        return 'lost at ht and drew ft'
    if ht_pos<0 and ft_pos>0:
        return 'lost at ht and won at ft'
    if ht_pos<0 and ft_pos<0:
        return 'lost at ht and ft'
df['away_ht_ft']=df.apply(lambda r: away_pos(r), axis=1)
```

```python
def home_double_chance(r):
    ht_pos=r['htHG']-r['htAG']
    sec_half_home=r['FTHG']-r['htHG']
    sec_half_away=r['FTAG']-r['htAG']
    sec_half_home_pos=sec_half_home-sec_half_away
    if ht_pos>0 and sec_half_home_pos>0:
        return 'won both halves'
    elif ht_pos>0 or sec_half_home_pos>0:
        return 'won either half'
    else:
        return 'didnt win either half'
    
df['home_double_chance']=df.apply(lambda r:  home_double_chance(r), axis=1)
```

```python
def away_double_chance(r):
    ht_pos=r['htAG']-r['htHG']
    sec_half_home=r['FTHG']-r['htHG']
    sec_half_away=r['FTAG']-r['htAG']
    sec_half_away_pos=sec_half_away-sec_half_home
    if ht_pos>0 and sec_half_away_pos>0:
        return 'won both halves'
    elif ht_pos>0 or sec_half_away_pos>0:
        return 'won either half'
    else:
        return 'didnt win either half'
    
df['away_double_chance']=df.apply(lambda r:  away_double_chance(r), axis=1)
```
